# Read in the files (as they are already generated)

```python
df_ground_k2_28b = pd.read_csv('../data/machine_readable/k2_28b_ground_final.txt',sep="\t",comment="#")
df_ground_k2_100b = pd.read_csv('../data/machine_readable/k2_100b_ground_final.txt',sep="\t",comment="#")
```

```python
# Re-create the light curves and the best fit models using the plot_transits function and the machine-readable tables
import plot_transits
import matplotlib.pylab as plt
%matplotlib inline

#Plot K2-28b light curve with plot_transits function
fig, ax = plt.subplots(figsize=(12,8))
plot_transits.plot_transit_with_model(df_ground_k2_28b['BJD_TDB'],
                        df_ground_k2_28b['rel_flux'],
                        df_ground_k2_28b['rel_flux_err'],
                        yresidual=df_ground_k2_28b['residual'],
                        xmodel=df_ground_k2_28b['model_time'],
                        ymodel=df_ground_k2_28b['model_flux'],ax=ax,offset=0.98)
ax.set_title('K2-28b')



#Plot K2-100b light curve with plot_transits function
fig, ax1 = plt.subplots(figsize=(12,8))
plot_transits.plot_transit_with_model(df_ground_k2_100b['BJD_TDB'],
                        df_ground_k2_100b['rel_flux'],
                        df_ground_k2_100b['rel_flux_err'],
                        yresidual=df_ground_k2_100b['residual'],
                        xmodel=df_ground_k2_100b['model_time'],
                        ymodel=df_ground_k2_100b['model_flux'],ax=ax1,offset=0.99)
ax1.set_title('K2-100b')
```
